## matplotlib

```python
import matplotlib.pyplot as plt

# Simple scatter plot
plt.scatter(df['Impression_Rating'], df['Hire_Rating'])
plt.xlabel('Impression Rating')
plt.ylabel('Hire Rating')
plt.title('Impression vs. Hire Rating')
plt.show()

# Histogram
plt.hist(df['Intellect_Rating'], bins=10)
plt.show()
```

## seaborn

```python
# Distribution plot
sns.histplot(data=df, x='Intellect_Rating')
```

## scipy

```python
from scipy import stats

# create groups
audio_group = df[df['CONDITION'] == 1]['Intellect_Rating']
transcript_group = df[df['CONDITION'] == 0]['Intellect_Rating']

# T-test
t_stat, p_value = stats.ttest_ind(audio_group, transcript_group)
print(f"Independent samples t-test results:")
print(f"  t({len(audio_group) + len(transcript_group) - 2}) = {t_stat:.2f}")
print(f"  p = {p_value:.3f}")
```

## statsmodels

```python
import statsmodels.formula.api as smf

# Linear regression
model = smf.ols('Hire_Rating ~ Impression_Rating + Intellect_Rating', data=df)
results = model.fit()
print(results.summary())
```
